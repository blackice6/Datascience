```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# use the describe function
data.describe()
```

```python
# we have noticed that the collumn item price collumn contains three empty slots 
# we need to fill this empty slots with a value and i  this case we shall fill with the min of the collumn

data["itemprice"] = data["itemprice"].fillna(55)


#describe the data again 
data.describe()
```

```python
# get the number of entries in a given collumn
data.groupby("gender").size()
```

```python
# combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# below is a count plot showing the distribution of males to females
sns.countplot(x="gender", data=data, palette="dark")
plt.title("Gender distribution against total count")
plt.ylabel("Total count of individuals")
```

```python

data.groupby("product").size()
```

```python
# Class Work: Create a Count Plot of product
# Which product category appears the most?
sns.countplot(x="product", data=data, palette="dark")
plt.xticks( rotation = 45)
plt.title("distribution of products")
```

```python
# Below is a countplot that has two coloumns
sns.countplot(x="gender", hue="city", palette="Reds", data=data)
```

```python
# below is a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
